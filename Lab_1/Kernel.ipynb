```python
import numpy as np
import pandas as pd
import math 
```

```python
x_values = np.linspace(-5, 5, 100000)
kernels = ['Rectangular', 'Triangular', 'Parabolic', 'Beweigh', 'Gaussian', 'Silverman']
```

```python
e = math.e
pi = math.pi
```

```python


# Hàm Y(x)
def Y(x: float):
    return 1 if -1 <= x <= 1 else 0

# Hàm Kernel
def Kernel(x: float, kernel_type: str):
    pi = math.pi
    e = math.e
    
    Kernel_dict = {
        'Rectangular': 0.5 * Y(x),
        'Triangular': max(0, 1 - abs(x)) * Y(x),
        'Parabolic': max(0, (3 / 4) * (1 - x**2)) * Y(x),
        'Beweigh': max(0, (15 / 16) * (1 - x**2)**2) * Y(x),
        'Gaussian': ((1 / (2 * pi)**0.5) * e**(-0.5 * x**2)) * Y(x),
        'Silverman': ((1 / 2) * e**(-abs(x) / (2**0.5)) * math.sin(abs(x) / (2**0.5) + (pi / 4))) * Y(x)
    }
    
    return Kernel_dict.get(kernel_type, None)

```

```python

kernel_results = {kernel: [Kernel(x, kernel) for x in x_values] for kernel in kernels}

df = pd.DataFrame(kernel_results, index=x_values)

print(df.head())
```

```python
import matplotlib.pyplot as plt

df.plot(title="Kernel Functions (x ∈ [0,1])")
plt.xlabel("x")
plt.ylabel("Kernel Value")
plt.legend(title="Kernel Type")
plt.show()
```
